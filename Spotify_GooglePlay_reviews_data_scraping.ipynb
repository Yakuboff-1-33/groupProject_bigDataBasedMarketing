```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.spotify.music',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=60000, 
    filter_score_with=None
)

result, _ = reviews(
    'com.spotify.music',
    continuation_token=continuation_token
)
```

```python
result
```

```python
for i in result:
  print(i['at'], i['content'], i['score'], i['thumbsUpCount'], i['replyContent'], i['userName'])
```

```python
import pandas as pd

time_submitted = []
review = []
rating = []
author_account_name = []
reply = []

for a in result:
    time_submitted.append(a['at'])
    review.append(a['content'])
    rating.append(a['score'])
    author_account_name.append(a['userName'])
    reply.append(a['replyContent'])

print(time_submitted)
print(review)
print(rating)
print(author_account_name)
print(reply)

df = pd.DataFrame({'time_submitted': time_submitted, 'review': review, 'rating': rating, 'author_account_name': author_account_name, 'reply': reply})
df.to_csv('reviews_newest_top60k.csv', index=False, encoding='utf-8')
```
